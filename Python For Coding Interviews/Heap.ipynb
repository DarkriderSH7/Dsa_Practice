```python
import heapq
from typing import List


def heap_pop(heap: List[int]) -> List[int]:
    result=[]
    while heap:
        result.append(heapq.heappop(heap))
    return result

# do not modify below this line
print(heap_pop([1, 2, 3]))
print(heap_pop([1, 3, 2]))
print(heap_pop([6, 7, 8, 12, 9, 10]))

# Challenge
# Implement the function heap_pop(heap: List[int]) -> List[int] that pops all elements from the heap heap. The heap should be a min heap, meaning that the element with the smallest priority should be at index 0. After popping all elements, return them in a list.
```

```python
import heapq
from typing import List


def heapify_strings(strings: List[str]) -> List[str]:
    heapq.heapify(strings)
    return strings

def heapify_integers(integers: List[int]) -> List[int]:
    heapq.heapify(integers)
    return integers


def heap_sort(nums: List[int]) -> List[int]:
    heapq.heapify(nums)
    nums.sort()
    return nums


# do not modify below this line
print(heapify_strings(["b", "a", "e", "c", "d"]))
print(heapify_integers([3, 4, 5, 1, 2, 6]))
print(heap_sort([3, 4, 5, 1, 2, 6]))


# Challenge
# Implement the function heapify_strings(strings: List[str]) -> List[str] that takes a list of strings and returns a min heap of strings. The heap should be a min heap, meaning that the element with the smallest priority should be at index 0.
# Implement the function heapify_integers(integers: List[int]) -> List[int] that takes a list of integers and returns a min heap of integers. The heap should be a min heap, meaning that the element with the smallest priority should be at index 0.
# Implement the function heap_sort(nums: List[int]) -> List[int] that takes a list of integers and returns a sorted list of integers. The heap should be a min heap, meaning that the element with the smallest priority should be at index 0.
```

```python
import heapq
from typing import List


def get_reverse_sorted(nums: List[int]) -> List[int]:
    maxheap=[]
    heapq.heapify(nums)
    for i in nums:
        heapq.heappush(maxheap,-i)
    top=[]
    while maxheap:
        top.append(-heapq.heappop(maxheap))
    return top





# do not modify below this line
print(get_reverse_sorted([1, 2, 3]))
print(get_reverse_sorted([5, 6, 4, 2, 7, 3, 1]))
print(get_reverse_sorted([5, 6, -4, 2, 4, 7, -3, -1]))


# Challenge
# Implement the function get_reverse_sorted(nums: List[int]) -> List[int] that takes a list of integers and returns a sorted list of integers in descending order. The heap should be a max heap, meaning that the element with the largest priority should be at index 0.
# The function should use the heapq module to create a max heap and then pop all elements from the heap in descending order.
# The function should return the sorted list of integers in descending order.
# The function should not modify the original list of integers.
# The function should not use any additional data structures other than the heapq module.
```
